### **Problem Statement**
Given a username and password, check whether login is successful.

**Credentials:**
- username = "admin"
- password = "1234"

### **Requirements**
- Print **"Login Successful"** if both username and password match
- Otherwise print **"Invalid Credentials"**

### **Real-World Application**
Authentication Systems

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
username = "admin"
password = "1234" 

user_id = str(input("Enter username: ")) # user input (username)
pasw = str(input("Enter password: ")) # user input (password)

if(user_id == username and pasw == password): 
    print("Login Successful") # runs if both condition true
else:
    print("Invalid Credential")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    2Ô∏è‚É£ Pass / Fail Analyzer
</h3>

### **Problem Statement**
Analyze student results.

**Marks Data:**
- marks = [45, 78, 90, 33, 60]

### **Requirements**
- A student passes if marks ‚â• 50
- Count total number of pass students
- Count total number of fail students
- Print the final result clearly

### **Real-World Application**
Academic Evaluation Systems

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
marks = [45, 78, 90, 33, 60]
pass_count = 0 # pass counter
fail_count = 0 # fail counter
for i in marks:
    if(i >= 50):
        pass_count += 1
    else:
        fail_count +=1

print(f"Students Passed: {pass_count}")
print(f"Students Failed: {fail_count}")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    3Ô∏è‚É£ Simple Data Cleaner
</h3>

### **Problem Statement**
Clean and standardize user names.

**Input Data:**
- names = [" Alice ", "bob", " CHARLIE "]

### **Requirements**
- Remove extra spaces
- Convert all names to lowercase
- Print the cleaned list

### **Real-World Application**
Data Preprocessing & Cleaning

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>



```python
names = [" Alice ", "bob", " CHARLIE "]

for i, name in enumerate(names):
    name = name.strip().lower()
    names[i] = name

print(names)
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    4Ô∏è‚É£ Message Length Analyzer
</h3>

### **Problem Statement**
Analyze message sizes.

**Messages Data:**
- messages = ["Hi", "Welcome to the platform", "OK"]

### **Requirements**
- Print the length of each message
- Flag messages longer than 10 characters

### **Real-World Application**
Text Filtering & Validation Systems

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}")

    if length > 10: # flags if length > 10
        print("‚ùó Long message detected")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    5Ô∏è‚É£ Error Message Detector
</h3>

### **Problem Statement**
Detect error messages from system logs.

**Log Data:**
- logs = ["INFO", "ERROR", "WARNING", "ERROR"]

### **Requirements**
- Count the number of **"ERROR"** entries
- Print the total error count

### **Real-World Application**
Monitoring & Log Analysis Systems

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
err_count = 0 # counter
for log in logs: 
    if log == "ERROR":
        err_count += 1

print(f"Total Errors: {err_count}")
```
