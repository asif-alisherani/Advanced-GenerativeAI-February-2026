<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    1️⃣ Social Media – Post Engagement Analyzer
</h3>

### **Problem Statement**
Create a function to analyze engagement on a social media post.

### **Rules**

- Calculate total likes  
- If Total Likes ≥ 1000 → Viral Post  
- Otherwise → Normal Engagement  

### **Real-Time Use**
Social media analytics platforms, Marketing dashboards.

### **Hint**
- Use function  
- Use loop for calculations  

<h3 style="color: green;">⬇️ Solution</h3>

```python
def analyze_engagement(likes_list):
    total_likes = 0
    
    for likes in likes_list:
        total_likes += likes
    
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    return total_likes, status

post_likes = [250, 300, 150, 200, 150]
total, status = analyze_engagement(post_likes)

print(f"Total Likes: {total}, Post Status: {status}")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    3️⃣ Agriculture – Rainfall Adequacy Checker
</h3>

### **Problem Statement**
Create a function that analyzes seasonal rainfall data.

### **Rules**

- Calculate average rainfall  
- Average ≥ required level → Adequate Rainfall  
- Otherwise → Inadequate Rainfall  

### **Real-Time Use**
Climate monitoring systems, Irrigation planning tools.

### **Hint**
- Use function  
- Use loop and math operations  

<h3 style="color: green;">⬇️ Solution</h3>

```python
def check_rainfall_adequacy(rainfall_data, required_level):
    total_rainfall = 0
    
    for rainfall in rainfall_data:
        total_rainfall += rainfall
    
    average_rainfall = total_rainfall / len(rainfall_data)
    
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    return average_rainfall, status

seasonal_rainfall = [65, 80, 70, 75, 68] 
required = 70

avg_rainfall, status = check_rainfall_adequacy(seasonal_rainfall, required)

print(f"Average Rainfall: {int(avg_rainfall)}")
print(f"Rainfall Status: {status}")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    4️⃣ Social Media – Duplicate Account Detection
</h3>

### **Problem Statement**
Create a function to detect duplicate usernames in a system.

### **Rules**

- If duplicates exist → Flag accounts  
- Else → Accounts are unique  

### **Real-Time Use**
Platform moderation, Fraud detection systems.

### **Hint**
- Use function  
- Use set data structure  

<h3 style="color: green;">⬇️ Solution</h3>

```python
usernames = ["john_doe", "jane_smith", "john_doe", "mike_jones", "jane_smith"]

def find_duplicates(l):
    seen = set()
    dupes = set() 
    for item in l:
        if item in seen:
            dupes.add(item)
        else:
            seen.add(item)
    return dupes

print("Duplicates: ",find_duplicates(usernames))  
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    6️⃣ Agriculture – Premium Crop Price Filter
</h3>

### **Problem Statement**
Create a function that filters crop prices above a premium threshold.

### **Rules**

- Price > ₹2000 → Premium Crop  
- Else → Regular Crop  

### **Real-Time Use**
Market analytics, Export pricing systems.

### **Hint**
- Use function  
- Use list filtering logic  

<h3 style="color: green;">⬇️ Solution</h3>

```python
def filter_premium_crops(crop_prices, threshold=2000):
    premium_crops = []
    
    for price in crop_prices:
        if price > threshold:
            premium_crops.append(price)
    
    return premium_crops

crop_prices = [1500, 2500, 1800, 3200, 1900, 2200]
premium = filter_premium_crops(crop_prices)

print(f"Premium Crops: {premium}")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    9️⃣ E-Learning – Student Attendance Eligibility System
</h3>

### **Problem Statement**
Create a function to calculate attendance percentage and check exam eligibility.

### **Rules**

- Attendance ≥ 75% → Eligible  
- Attendance < 75% → Not Eligible  

### **Real-Time Use**
Learning management systems, Internship tracking platforms.

### **Hint**
- Use function  
- Use list and loop  

<h3 style="color: green;">⬇️ Solution</h3>

```python
def check_exam_eligibility(attendance_record):
    total_days = len(attendance_record)
    present_count = 0
    
    for status in attendance_record:
        if status == "P":
            present_count += 1
    
    attendance_percentage = (present_count / total_days) * 100
    
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    
    return attendance_percentage, eligibility

attendance = ["P", "P", "A", "P", "P", "P", "P", "A", "P", "P"]
percentage, eligibility = check_exam_eligibility(attendance)

print(f"Attendance Percentage: {percentage}")
print(f"Exam Eligibility: {eligibility}")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    🔟 Smart Electricity Bill Analyzer
</h3>

### **Problem Statement**
Create a function that calculates the electricity bill and classifies usage level.

### **Billing Rules**

- First 100 units → ₹3 per unit  
- Next 100 units (101–200) → ₹5 per unit  
- Above 200 units → ₹7 per unit  

### **Usage Classification**

- Total bill < ₹500 → Low Usage  
- ₹500 – ₹1500 → Moderate Usage  
- ₹1500 → High Usage  

### **Real-Time Use**
Smart meter systems, Electricity billing software, Energy analytics.

### **Hint**
- Use function to calculate bill  
- Use conditional logic for slabs  
- Use return values  

<h3 style="color: green;">⬇️ Solution</h3>

```python
def calculate_electricity_bill(units):
    bill = 0
    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Classify usage based on total bill
    if bill < 500:
        usage_level = "Low Usage"
    elif bill <= 1500:
        usage_level = "Moderate Usage"
    else:
        usage_level = "High Usage"
    
    return bill, usage_level


test_units = [80, 150, 300]
print("Test Cases")
for units in test_units:
    bill, usage = calculate_electricity_bill(units)
    print(f"\nUnits: {units}")
    print(f"Bill: ₹{bill}")
    print(f"Usage: {usage}")
```
