    1Ô∏è‚É£ Mobile Recharge Validation System
</h3>

### **Problem Statement**
Create a function to validate a mobile recharge request.

**Rules:**

- Recharge amount must be ‚â• ‚Çπ50  
- Valid plans: 199, 299, 399, 599  
- User should be allowed to retry if invalid  

### **Real-Time Use**
Telecom recharge apps, Payment gateways.

### **Hint**
- Use function for validation  
- Use while loop for retry  

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
def valid_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter Recharge Amount: ‚Çπ"))

        if amount < 50:
            print("Amount must be at least ‚Çπ50\n")

        elif amount not in valid_plans:
            print(f"Invalid plan. Choose from {valid_plans}\n")

        else:
            print(f"Recharge of ‚Çπ{amount} successful")
            break


valid_recharge()
```

    2Ô∏è‚É£ Inventory Reorder Alert System
</h3>

### **Problem Statement**
Given a dictionary of products and stock quantities:

- Print **Reorder Alert** if stock < 15  
- Print **Stock OK** otherwise  

### **Real-Time Use**
Warehouse systems, Retail management.

### **Hint**
- Use function  
- Use for loop on dictionary  

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
def check_inventory(inventory):
    for product, stock in inventory.items():
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock: {stock})")
        else:
            print(f"{product}: Stock OK (Stock: {stock})")

inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 8,
    "Monitor": 15,
    "Headphones": 5,
    "Webcam": 20
}

check_inventory(inventory)
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    3Ô∏è‚É£ Student Result Processing System
</h3>

### **Problem Statement**
Create a function that:

- Accepts a list of student marks  
- Calculates average marks  
- Prints **Pass** if average ‚â• 50 else **Fail**  

### **Real-Time Use**
Online exam systems, Evaluation portals.

### **Hint**
- Use loop to calculate total  
- Return result from function  

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
def calculate_result(marks):
    if len(marks) == 0:
        return "No marks provided", 0
    
    total = 0
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"
    
    return result, average

# Example usage
student_marks = [45, 67, 55, 72, 60]
result, avg = calculate_result(student_marks)

print(f"Student Marks: {student_marks}")
print(f"Average Marks: {avg:.2f}")
print(f"Result: {result}")

# Another example
student2_marks = [35, 42, 38, 45, 40]
result2, avg2 = calculate_result(student2_marks)

print(f"\nStudent 2 Marks: {student2_marks}")
print(f"Average Marks: {avg2:.2f}")
print(f"Result: {result2}")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    4Ô∏è‚É£ Cab Fare Estimator with Retry Option
</h3>

### **Problem Statement**
Create a cab fare calculator:

- Base fare = ‚Çπ50  
- ‚Çπ12 per km  
- Peak hour ‚Üí 25% extra  
- Allow retry using while loop  

### **Real-Time Use**
Ride-hailing apps, Transport systems.

### **Hint**
- Function for fare calculation  
- While loop for retry  

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
def calculate_fare(distance, is_peak_hour=False):
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    if is_peak_hour:
        peak_charge = fare * 0.25
        fare += peak_charge
        print(f"Peak hour surcharge (25%): ‚Çπ{peak_charge:.2f}")
    
    return fare

def cab_fare_system():
    while True:
        try:
            distance = float(input("Enter distance in km: "))
            peak_hour = input("Is it peak hour? (yes/no): ").lower()
            
            is_peak = peak_hour == 'yes'
            
            fare = calculate_fare(distance, is_peak)
            
            print(f"\n--- Fare Breakdown ---")
            print(f"Base Fare: ‚Çπ50")
            print(f"Distance: {distance} km @ ‚Çπ12/km")
            print(f"Total Fare: ‚Çπ{fare:.2f}")
            
            retry = input("\nCalculate another fare? (yes/no): ").lower()
            if retry != 'yes':
                print("Thank you for using our service!")
                break
                
        except ValueError:
            print("Please enter valid input!")

cab_fare_system()
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    5Ô∏è‚É£ Employee Attendance Eligibility Checker
</h3>

### **Problem Statement**
Create a function that:

- Accepts attendance list ("P", "A")  
- Calculates attendance percentage  
- Returns **Eligible** if ‚â• 75% else **Not Eligible**  

### **Real-Time Use**
HR payroll systems, Internship tracking.

### **Hint**
- Use loop to count present days  
- Use return statement  

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
def check_attendance_eligibility(attendance_list):
    if len(attendance_list) == 0:
        return "No attendance data", 0
    
    present_count = 0
    
    for day in attendance_list:
        if day == "P":
            present_count += 1
    
    attendance_percentage = (present_count / len(attendance_list)) * 100
    
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    
    return eligibility, attendance_percentage

employee1 = ["P", "P", "A", "P", "P", "P", "A", "P", "P", "P"]
eligibility1, percentage1 = check_attendance_eligibility(employee1)

print(f"Employee 1 Attendance: {employee1}")
print(f"Attendance Percentage: {percentage1:.2f}%")
print(f"Eligibility Status: {eligibility1}")

employee2 = ["P", "A", "A", "P", "A", "P", "A", "A", "P", "A"]
eligibility2, percentage2 = check_attendance_eligibility(employee2)

print(f"\nEmployee 2 Attendance: {employee2}")
print(f"Attendance Percentage: {percentage2:.2f}%")
print(f"Eligibility Status: {eligibility2}")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    6Ô∏è‚É£ Password Strength Checker
</h3>

### **Problem Statement**
Create a function to check password strength.

**Rules:**

- Minimum length = 8  
- Must contain at least one digit  
- Must contain at least one special character (@#$)  

### **Real-Time Use**
Login systems, Security validation.

### **Hint**
- Use loop over string  
- Use conditions inside function  

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
def check_password_strength(password):
    if len(password) < 8:
        return False, "Password must be at least 8 characters long"
    
    has_digit = False
    for char in password:
        if char.isdigit():
            has_digit = True
            break
    
    if not has_digit:
        return False, "Password must contain at least one digit"
    
    special_chars = "@#$"
    has_special = False
    for char in password:
        if char in special_chars:
            has_special = True
            break
    
    if not has_special:
        return False, "Password must contain at least one special character (@#$)"
    
    return True, "Strong password!"

def password_validator():
    while True:
        password = input("Enter password: ")
        
        is_strong, message = check_password_strength(password)
        print(message)
        
        if is_strong:
            print("Password accepted!")
            break
        
        retry = input("Try again? (yes/no): ").lower()
        if retry != 'yes':
            print("Password setup cancelled.")
            break

password_validator()
```
