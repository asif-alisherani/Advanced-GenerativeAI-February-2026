<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    1️⃣ Smart Parking Lot Management System
</h3>

### **Problem Statement**
Design a function to manage a smart parking lot.

The system should:

- Accept vehicle entry and exit logs  
- Calculate total parked vehicles  
- Identify peak parking usage  
- Alert if parking exceeds capacity  

### **Real-World Application**
Mall parking systems, Smart city infrastructure.

### **Hint**
- Function + list  
- Loop for counting  
- Conditional alert  

**Input:**
Parking Capacity: 50  
Vehicle Logs: `["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]`

<h3 style="color: green;">⬇️ Solution</h3>

```python
def manage_parking_lot(capacity, vehicle_logs):
    currently_parked = 0
    
    for log in vehicle_logs:
        if log == "IN":
            currently_parked += 1
        elif log == "OUT":
            if currently_parked > 0:
                currently_parked -= 1
    
    if currently_parked >= capacity:
        status = "Full"
    else:
        status = "Available"
    
    return currently_parked, status

parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

currently_parked, status = manage_parking_lot(parking_capacity, vehicle_logs)

print(f"Currently Parked Vehicles: {currently_parked}")
print(f"Parking Status: {status}")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    2️⃣ Online Food Delivery Time Estimator
</h3>

### **Problem Statement**
Create a function that estimates delivery time based on:

- Distance (km)  
- Weather condition  
- Traffic level  

Apply delays dynamically and display final ETA.

### **Real-World Application**
Food delivery apps, Logistics platforms.

### **Hint**
- Function + multiple conditions  
- Mathematical adjustments  

**Input:**
Distance (km): 8  
Traffic Level: High  
Weather: Rainy  

<h3 style="color: green;">⬇️ Solution</h3>

```python
def estimate_delivery_time(distance, traffic_level, weather):
    base_time = distance * 3
    
    if traffic_level == "High":
        base_time += 15
    elif traffic_level == "Medium":
        base_time += 8
    
    if weather == "Rainy":
        base_time += 12
    elif weather == "Stormy":
        base_time += 20
    
    # Preparation time
    base_time += 10
    
    return int(base_time)

distance = 8
traffic = "High"
weather = "Rainy"

eta = estimate_delivery_time(distance, traffic, weather)

print(f"Estimated Delivery Time: {eta} minutes")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    3️⃣ Movie Theatre Seat Occupancy Analyzer
</h3>

### **Problem Statement**
Build a function that analyzes seat booking data and:

- Calculates occupancy percentage  
- Determines if show is Housefull  
- Suggests opening additional shows  

### **Real-World Application**
Cinema ticketing systems, Event management software.

### **Hint**
- Function + list  
- Percentage calculation  

**Input:**
Total Seats: 200  
Booked Seats: (150 bookings)

<h3 style="color: green;">⬇️ Solution</h3>

```python
def analyze_seat_occupancy(total_seats, booked_seats):
    num_booked = 0
    for seat in booked_seats:
        if seat == 1:
            num_booked += 1
    
    occupancy = (num_booked / total_seats) * 100
    
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Available"
    
    return occupancy, status

total_seats = 200
booked_seats = [1] * 150  # 150 booked seats

occupancy, status = analyze_seat_occupancy(total_seats, booked_seats)

print(f"Occupancy: {int(occupancy)}%")
print(f"Show Status: {status}")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    4️⃣ Cloud Server Load Classification System
</h3>

### **Problem Statement**
Create a function to classify server load based on CPU usage readings.

### **Rules**

- Average CPU < 50% → Normal  
- 50% – 80% → Warning  
- Above 80% → Critical  

### **Real-World Application**
Cloud monitoring dashboards, DevOps alerting systems.

### **Hint**
- Function + loop  
- Average calculation  

**Input:**
CPU Readings (%): `[45, 60, 70, 85, 90]`

<h3 style="color: green;">⬇️ Solution</h3>

```python
def classify_server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading
    
    average = total / len(cpu_readings)
    
    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"
    
    return average, status

cpu_readings = [45, 60, 70, 85, 90]

average, status = classify_server_load(cpu_readings)

print(f"Average CPU Load: {int(average)}%")
print(f"Server Status: {status}")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    5️⃣ Smart Classroom Resource Usage Monitor
</h3>

### **Problem Statement**
Design a function that tracks usage of classroom resources and identifies overuse patterns.

### **Real-World Application**
Smart classrooms, Energy optimization systems.

### **Hint**
- Function + dictionary  
- Conditional checks  

**Input:**
Resource Usage (hours):  
{
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

<h3 style="color: green;">⬇️ Solution</h3>

```python
def monitor_classroom_resources(resource_usage, threshold=8):
    overused = []
    
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)
    
    if len(overused) > 0:
        alert = "Yes"
    else:
        alert = "No"
    
    return overused, alert

resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

overused, alert = monitor_classroom_resources(resource_usage)

print(f"Overused Resources: {', '.join(overused)}")
print(f"Energy Alert: {alert}")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    6️⃣ Online Event Registration Capacity Controller
</h3>

### **Problem Statement**
Create a function that manages event registrations by:

- Tracking registrations  
- Preventing overbooking  
- Triggering waitlist mode  

### **Real-World Application**
Webinar platforms, Conference registration systems.

### **Hint**
- Function + loop  
- Capacity validation  

**Input:**
Event Capacity: 100  
Registrations: 105  

<h3 style="color: green;">⬇️ Solution</h3>

```python
def manage_event_registration(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    
    return confirmed, waitlisted, status

event_capacity = 100
registrations = 105

confirmed, waitlisted, status = manage_event_registration(event_capacity, registrations)

print(f"Confirmed Registrations: {confirmed}")
print(f"Waitlisted Users: {waitlisted}")
print(f"Registration Status: {status}")
```
