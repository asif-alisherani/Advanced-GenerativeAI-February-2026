<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    1Ô∏è‚É£ Unique Words in a Sentence
</h3>

### **Problem Statement**
Given a sentence, identify all unique words and display:

- The count of unique words  
- The unique words themselves  

**Input:**  
`python is easy and python is powerful`

### **Real-World Application**
Text analytics, NLP preprocessing, Search engine indexing.

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
sentence = "python is easy and python is powerful"

words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)    
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    2Ô∏è‚É£ Highest Salary from Employee Data
</h3>

### **Problem Statement**
Store employee names and their salaries, then determine which employee has the highest salary.

**Input:**
employees = {  
&nbsp;&nbsp;&nbsp;&nbsp;"Ravi": 75000,  
&nbsp;&nbsp;&nbsp;&nbsp;"Anita": 68000,  
&nbsp;&nbsp;&nbsp;&nbsp;"Kiran": 72000  
}

### **Real-World Application**
HR management systems, Payroll analysis.

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000,
    "Rahul": 45000,
    "Monika": 51900,
    "Ali": 92000
}
high = 0
person = ""
for emp in employees:
    if employees[emp] > high:
        person = emp
        high = employees[emp]
print(f"Highest salary: {person} - {high}")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    3Ô∏è‚É£ Find Maximum and Minimum Values
</h3>

### **Problem Statement**
Find the maximum and minimum values from a list of numbers.

**Input:**  
`numbers = [45, 22, 89, 10, 66]`

### **Real-World Application**
Sales data analysis, Temperature monitoring, Performance evaluation.

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
numbers = [45,22,89,10,66,29,43,16]
maximum = minimum = numbers[0]
for i in numbers:
    if i < minimum:
        minimum = i
    elif i > maximum:
        maximum = i
print("List:",numbers)
print("Max:",maximum)
print("Min:",minimum)
```

    5Ô∏è‚É£ Calculate Attendance Percentage
</h3>

### **Problem Statement**
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

**Input:**  
`attendance = ["P", "P", "A", "P", "P"]`

### **Real-World Application**
Attendance tracking, HR management systems.

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
attendance = ["P", "P", "A", "P", "P"]
pcount = 0
total = 0
for i in attendance:
    total += 1
    if i == "P":
        pcount += 1
print("Attendance Percentage:", (pcount/total)*100)
```

    7Ô∏è‚É£ Count Character Frequency
</h3>

### **Problem Statement**
Count how many times each character appears in a string.

**Input:**  
`python`

### **Real-World Application**
Text analysis, Password validation.

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
string = "python"
chars = {}
for i in string:
    if i in chars:
        chars[i] += 1
    else:
        chars[i] = 1
print(chars)
```
