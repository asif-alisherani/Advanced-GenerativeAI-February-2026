<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    1Ô∏è‚É£ Employee Performance Bonus Eligibility
</h3>

### **Problem Statement**
A company evaluates employee performance scores at the end of the year. You are given a dictionary containing employee names and their performance scores.

**Input:**
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

### **Requirements**
- Identify the highest performance score  
- Handle ties if multiple employees have the same highest score  
- Display all employees eligible for the top performance bonus  

### **Real-World Application**
HR Performance Evaluation Systems  

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Highest performance score
highest_score = max(employees.values())

# Employees with the highest score
top_performers = [name for name, score in employees.items() if score == highest_score]

print("Top Performers Eligible for Bonus:", 
      ", ".join(top_performers), 
      f"(Score: {highest_score})")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    2Ô∏è‚É£ Search Query Keyword Analysis
</h3>

### **Problem Statement**
An e-commerce website stores customer search queries. You are given a search query sentence entered by a user.

**Input:**
"Buy mobile phone buy phone online"

### **Requirements**
- Convert the input to lowercase  
- Ignore common punctuation  
- Count the frequency of each keyword  
- Display only keywords searched more than once  

### **Real-World Application**
Search Engine Optimization (SEO) & Keyword Analytics  

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
query = "Buy mobile phone buy phone online"
# Convert to lowercase and remove common punctuation
query = query.lower()
for char in ".,!?;:":
    query = query.replace(char, "")

# Split into words
words = query.split()

# Frequency of each word
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# Filter words that appear more than once
repeated_words = {word: count for word, count in word_count.items() if count > 1}

print(repeated_words)
```

    3Ô∏è‚É£ Sensor Data Validation
</h3>

### **Problem Statement**
A factory collects sensor readings every hour. Each reading is stored in a list where the index represents the hour and the value represents the sensor reading.

**Input:**
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

### **Requirements**
- Identify readings that are even numbers (valid readings)  
- Store them as (hour_index, reading_value) pairs  
- Ignore odd readings (invalid readings)  

### **Real-World Application**
Industrial Monitoring Systems  

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
# Filter even readings and store as (index, value) tuples
valid_readings = [(index, value) for index, value in enumerate(sensor_readings) 
                  if value % 2 == 0]

print(valid_readings)
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    4Ô∏è‚É£ Email Domain Usage Analysis
</h3>

### **Problem Statement**
A company wants to analyze which email providers its users are using. You are given a list of employee email IDs.

**Input:**
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

### **Requirements**
- Count how many users belong to each email domain  
- Calculate the percentage usage of each domain  

### **Real-World Application**
User Demographics & Platform Analytics  

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
emails = [ "ravi@gmail.com", "anita@yahoo.com", "kiran@gmail.com", "suresh@gmail.com", "meena@yahoo.com" ]
# Count domains
domain_count = {}
for email in emails:
    # Extract domain (part after @)
    domain = email.split('@')[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculate total emails
total_emails = len(emails)

# Calculate and display percentages
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage}%")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    5Ô∏è‚É£ Sales Spike Detection
</h3>

### **Problem Statement**
A retail company tracks daily sales. Sudden spikes in sales may indicate promotions or unusual activity.

**Input:**
sales = [1200, 1500, 900, 2200, 1400, 3000]

### **Requirements**
- Calculate the average daily sales  
- Detect days where sales are more than 30% above average  
- Display the day number and sale value  

### **Real-World Application**
Business Intelligence & Sales Monitoring  

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
# Average sales
average_sales = sum(sales) / len(sales)

# Calculate threshold (30% above average)
threshold = average_sales * 1.30

# Detect and display spike days
for day, sale in enumerate(sales, start=1):
    if sale > threshold:
        print(f"Day {day}: {sale}")
```

<h3 style="background: white; color: blue; text-align: center; padding: 15px;">
    6Ô∏è‚É£ Duplicate User ID Detection
</h3>

### **Problem Statement**
A system stores user IDs during registration. Duplicate IDs can cause data integrity issues.

**Input:**
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

### **Requirements**
- Identify duplicate user IDs  
- Display how many times each duplicate appears  

### **Real-World Application**
Database Integrity & User Management Systems  

<h3 style="color: green;">‚¨áÔ∏è Solution</h3>


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
# Occurrences of each user ID
id_count = {}
for user_id in user_ids:
    id_count[user_id] = id_count.get(user_id, 0) + 1

# Display only duplicates (count > 1)
for user_id, count in id_count.items():
    if count > 1:
        print(f"{user_id} ‚Üí {count} times")
```
